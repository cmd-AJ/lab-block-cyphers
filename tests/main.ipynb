```python

key = generate_3des_key(2) #En este caso utilizamos 2 como 16 bytes
iv = generate_iv()

plaintext = b"Mensaje secreto super 3DES"

#Declaración del variable en base a las funciones 
ciphertext = encrypt_3des_cbc(plaintext, key, iv)
print("Texto cifrado usando 3des:", ciphertext)


#Funcion para poder desencriptar 3des utilizado en la parte de arriba
decrypted = decrypt_3des_cbc(ciphertext, key, iv)
print("Texto que se necesita desencriptar:", decrypted)

print("Es lo mismo?: ", decrypted == plaintext)
```

```python
msg = "ATAQUE ATAQUE ATAQUE ATAQUE"

ct_ecb, _      = encrypt_text(msg, key, "ECB")
ct_cbc, iv_cbc = encrypt_text(msg, key, "CBC")

print("ECB cipher (hex):", ct_ecb.hex())
print("CBC cipher (hex):", ct_cbc.hex())

# ECB: repeated plaintext blocks → repeated ciphertext blocks (visible pattern)
# CBC: chaining breaks the repetition → no visible pattern

print("ECB decrypted:", decrypt_text(ct_ecb, key, "ECB"))
print("CBC decrypted:", decrypt_text(ct_cbc, key, "CBC", iv=iv_cbc))
```
