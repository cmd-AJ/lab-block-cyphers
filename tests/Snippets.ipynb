```python
msg1 = b"ABCDE"   # Mensaje de 5 bytes
padded1 = pkcs7_pad(msg1, 8)
print("Original:", msg1)
print("Padded (hex):", padded1.hex())
print("Unpadded:", pkcs7_unpad(padded1))
```

"ABCDE"
Faltan 3 bytes para completar el bloque de 8. Por lo tanto se agregan 3 bytes con valor 0x03.03 03 03 indica que se agregaron 3 bytes.

```python
msg2 = b"12345678"   # Mensaje de 8 bytes
padded2 = pkcs7_pad(msg2, 8)
print("Original:", msg2)
print("Padded (hex):", padded2.hex())
print("Unpadded:", pkcs7_unpad(padded2))
```

El mensaje ya es múltiplo de 8 entonces se tiene que agregar un bloque completo adicional. Por lo tanto, se agregan 8 bytes con valor 0x08.

```python
msg3 = b"ABCDEFGHIJ"   # Mensaje de 10 bytes
padded3 = pkcs7_pad(msg3, 8)
print("Original:", msg3)
print("Padded (hex):", padded3.hex())
print("Unpadded:", pkcs7_unpad(padded3))
```

Para "ABCDEFGHIJ" le faltan 6 bytes para completar el bloque Entonces debemos de agregar 6 bytes con valor 0x06.
